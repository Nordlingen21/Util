```python
import random
```

```python
def NumAleatorio():
    a=random.random()
    a=a*10
    a=int(a+1)
    return a
```

```python
def preguntar(aciertos,errores):
    a=NumAleatorio()
    b=NumAleatorio()
    c=a*b
    print('')
    print(str(a)+' x '+str(b))
    respuesta=input()

    if int(respuesta)==c:
        print(str(a)+' x '+str(b)+' = '+str(c)+' FENOMENAL!!!')
        aciertos=aciertos+1
    else:
        print('Melón, te has equivocado!!!!')
        print(str(a)+' x '+str(b)+' = '+str(c))
        errores=errores+1
    
    print('Aciertos='+str(aciertos))
    print('Errores='+str(errores))
    print('Preguntas='+str(aciertos+errores))
    
    return aciertos, errores
    
```
